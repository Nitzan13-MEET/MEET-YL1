```python
import csv
f = open("cig.csv", "rb")
r = csv.reader(f)

tar=[]
n=[]

for row in r:
    x=row[0].replace("\t", " ")
    z=x.split()
    z=z[1:]
    s=map(float,z)
    tar.append(s[0])
    n.append(s[1])

print tar
print n
    
    
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(tar,n)
ax.set_title('Nicotine and Tar')
ax.set_xlabel('Tar')
ax.set_ylabel('Nicotine')
savefig("N_and_T.png")
```
